```python
import pandas as pd
import os
import matplotlib.pylab as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import sklearn.metrics
import statsmodels.formula.api as smf

#Load the dataset

data_clean1 = pd.read_csv('NewDataset.csv')

c5 = data_clean1["Profit"].value_counts(sort = False)
print(data_clean1["Profit"])
print(c5)


# convert to numeric format
data_clean1['Isweekday'] = pd.to_numeric(data_clean1['Isweekday'], errors='coerce')
data_clean1['Isweekend'] = pd.to_numeric(data_clean1['Isweekend'], errors='coerce')
data_clean1['Isworkday'] = pd.to_numeric(data_clean1['Isworkday'], errors='coerce')
data_clean1['IsBankHoliday'] = pd.to_numeric(data_clean1['IsBankHoliday'], errors='coerce')
data_clean1['ProductSalePrice'] = pd.to_numeric(data_clean1['ProductSalePrice'], errors='coerce')
data_clean1['OrderItemQuantity'] = pd.to_numeric(data_clean1['OrderItemQuantity'], errors='coerce')
data_clean1['PurchasePrice'] = pd.to_numeric(data_clean1['PurchasePrice'], errors='coerce')
data_clean1['Profit'] = pd.to_numeric(data_clean1['Profit'], errors='coerce')


# bivariate bar graph to visualize the relationship between profit and the predictor variables
#ProductSalePrice vs Profit
sns.factorplot(x="ProductSalePrice", y="Profit", data=data_clean1, kind="bar", ci=None, size=10)
plt.xlabel('ProductSalePrice')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph IsweekendRC vs Profit 
sns.factorplot(x="Isweekend", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('Isweekend')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph IsweekdayRC vs Profit
sns.factorplot(x="Isweekday", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('Isweekday')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph IsweekdayRC vs Profit
sns.factorplot(x="Isworkday", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('Isworkday')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph IsBankHolidayRC vs Profit
sns.factorplot(x="IsBankHoliday", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('IsBankHoliday')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph ProductSalePrice vs Profit
sns.factorplot(x="OrderItemQuantity", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('OrderItemQuantity')
plt.ylabel('Profit')
plt.show()

# bivariate bar graph SupPurchasePrice vs Profit
sns.factorplot(x="PurchasePrice", y="Profit", data=data_clean1, kind="bar", ci=None, size=5)
plt.xlabel('PurchasePrice')
plt.ylabel('Profit')
plt.show()

#check mean, standard deivation and maximum ,adn min values of quantitative varibles
print(data_clean1['ProductSalePrice'].describe())
print(data_clean1['OrderItemQuantity'].describe())
print(data_clean1['PurchasePrice'].describe())

#ProductSalePrice run the  scatterplots together to get both linear and 2rd order fit lines for x121_2012
scat1 = sns.regplot(x="ProductSalePrice", y="Profit", scatter=True, order=2, data=data_clean1)
plt.xlabel('ProductSalePrice')
plt.ylabel('Profit')
plt.show()

##the resultant relationship from the scatterplot is not linear so i will test significans using regression
# center quantitative IVs for regression analysis
data_clean1['ProductSalePrice_c'] = (data_clean1['ProductSalePrice'] - data_clean1['ProductSalePrice'].mean())
data_clean1['OrderItemQuantity_c'] = (data_clean1['OrderItemQuantity'] - data_clean1['OrderItemQuantity'].mean())
data_clean1['PurchasePrice_c'] = (data_clean1['PurchasePrice'] - data_clean1['PurchasePrice'].mean())
data_clean1[["ProductSalePrice_c", "OrderItemQuantity_c", "PurchasePrice_c"]].describe()

# linear regression analysis
reg1 = smf.ols('Profit ~ ProductSalePrice_c', data=data_clean1).fit()
print(reg1.summary())

#including all other predictors
predictors = data_clean1[['Isweekday','Isweekend','Isworkday','IsBankHoliday',
'ProductSalePrice','PurchasePrice','OrderItemQuantity']]

print(data_clean1.dtypes)

targets = data_clean1.Profit

x_train, x_test, y_train, y_test = train_test_split(predictors, targets, test_size=0.2, random_state=5)

print(x_train.shape)
print(x_test.shape)

print(y_train.shape)
print(y_test.shape)

print('************* Decision Tree *****************')
#Build model on training data
classifier = DecisionTreeRegressor()
classifier.fit(x_train,y_train)
predictions=classifier.predict(x_test)


from sklearn.metrics import r2_score

print('R2Socre: ',r2_score(y_test, predictions))
# print(classifier.score(x_train, y_train))

import joblib

joblib.dump(classifier, 'price_estimator.pkl')
```
